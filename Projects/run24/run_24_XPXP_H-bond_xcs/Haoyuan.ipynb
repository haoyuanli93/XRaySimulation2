# Water

```python
import numpy as np
import matplotlib.pyplot as plt
# Define target delay filter
target_delays = np.concatenate((np.arange(0.0,0.5,0.05), np.arange(0.5,3.5,0.1), [9.376401551718187]))


data = np.load('./Water.npz',allow_pickle=True)

# Access the arrays using the correct keys
iso_q_lines_temp = data['iso_q_lines_temp']
Delta_I = data['Delta_I']

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming iso_q_lines and target_delays are defined, and averaged_q_by_delay is a dictionary

delays = np.array(target_delays)
delays=delays[:-1]
# Create a 2D array of intensity values
intensity_values = Delta_I[:-1,1:]

plt.figure(figsize=[6,4.5])
plt.imshow(intensity_values, 
           aspect='auto', 
           extent=[iso_q_lines_temp[0], iso_q_lines_temp[-1], 0, len(delays) - 1],
           vmin=0,vmax=10,
           cmap='viridis', 
           origin='lower')

plt.yticks(list(range(len(delays))), labels=['{:.1f} ps'.format(val) for val in delays])
plt.xlabel(r'Q [$\AA^{-1}$]')
plt.ylabel('Delay [ps]')
plt.xlim([0.04, 0.3])
plt.title('H2O')
plt.colorbar(label=r'$\Delta I(Q)$ [A.U.]')
plt.tight_layout()

plt.show()
```

```python

colors = plt.cm.plasma(np.linspace(0,1,11))
interdelay=[0,4,8,12,16,20,24,28,32,36,39]
target_delayss=target_delays[interdelay]
plt.figure(figsize=[5.5,4.5])
for i, delay in enumerate(target_delayss):
    plt.plot(iso_q_lines_temp[3:], Delta_I[interdelay[i], 3:], label='{:.2f} ps'.format(delay), color=colors[i])
    
plt.xlabel(r'Q [$\AA^{-1}$]')
plt.ylabel(r'$\Delta I(Q)$ [A.U.]')
plt.legend(title = 'Delay',loc = 'upper right')
plt.xlim([0.05,0.3])
plt.ylim([-1,8])

plt.title('H2O')

```

# CO2

```python
import numpy as np
import matplotlib.pyplot as plt
# Define target delay filter
target_delays = np.concatenate((np.arange(0.0,2.1,0.1), [2.9989, 3.9985, 4.99813, 5.9977, 9.996, 13.994]))

data = np.load('./CO2.npz',allow_pickle=True)

iso_q_lines=data ['iso_q_lines']
averaged_q_by_delay=data ['averaged_q_by_delay'].item()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming iso_q_lines and target_delays are defined, and averaged_q_by_delay is a dictionary
iso_q_lines_temp = np.array(iso_q_lines[:-1])  # convert to numpy array if not already
delays = np.array(target_delays)

# Create a 2D array of intensity values
intensity_values = np.array([averaged_q_by_delay[delay][:len(iso_q_lines_temp)] for delay in delays])

plt.figure()
plt.imshow(intensity_values, 
           aspect='auto', 
           extent=[iso_q_lines[0], iso_q_lines_temp[-1], 0, len(delays) - 1],
           vmax=150,
           cmap='viridis', 
           origin='lower')

plt.yticks(list(range(len(target_delays))), labels=['{:.1f} ps'.format(val) for val in target_delays])
plt.xlabel(r'Q [$\AA^{-1}$]')
plt.ylabel('Delay [ps]')
plt.xlim([0.03, 0.3])
plt.colorbar(label=r'$\Delta I(Q)$ [A.U.]')
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming iso_q_lines and target_delays are defined, and averaged_q_by_delay is a dictionary
iso_q_lines_temp = np.array(iso_q_lines[:-1])  # convert to numpy array if not already
delays = np.array(target_delays)

# Create a 2D array of intensity values
intensity_values = np.array([averaged_q_by_delay[delay] for delay in delays])

# Define the specific Q ranges
q_ranges = [0.04, 0.06, 0.08, 0.1, 0.12, 0.14, 0.16, 0.18, 0.20, 0.22, 0.24, 0.26]

plt.figure()

# Iterate through each Q range and calculate the average intensity
for i in range(len(q_ranges) - 1):
    q_min = q_ranges[i]
    q_max = q_ranges[i + 1]
    
    # Find the indices of the Q values within the current range
    q_indices = np.where((iso_q_lines_temp >= q_min) & (iso_q_lines_temp <= q_max))[0]
    
    # Calculate the average intensity over the specified Q range for each delay
    if len(q_indices) > 0:
        average_intensity = np.mean(intensity_values[:, q_indices], axis=1)
        
        # Plot the average ΔI vs. time delay
        plt.plot(delays, average_intensity, label=f'{q_min} ≤ Q ≤ {q_max} $\AA^{-1}$')

plt.xlabel('Delay [ps]')
plt.ylabel(r'$\Delta I(Q)$ [A.U.]')
plt.title(r'Average $\Delta I(Q)$ vs. Time Delay for Q Ranges')
plt.legend()
plt.grid(True)
plt.show()
```
