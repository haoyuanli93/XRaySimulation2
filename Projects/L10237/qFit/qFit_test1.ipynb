```python
def Rodrigue_rot(v, axis, rot_angle):
    if v.shape != (3,):
        v_rot = np.zeros_like(v)
        n1, n2, _ = v.shape

        for i in tqdm(range(n1)):
            for j in range(n2):
                vv = v[i,j]
                v_rot[i,j] = vv * np.cos(rot_angle) + np.cross(axis, vv) * np.sin(rot_angle) + axis * np.dot(axis, vv) * (1-np.cos(rot_angle))
        return v_rot
    else:
        return v * np.cos(rot_angle) + np.cross(axis, v) * np.sin(rot_angle) + axis * np.dot(axis, v) * (1-np.cos(rot_angle))

def calculate_rotation_matrix(angle_yaw=0, angle_pitch=0, angle_roll=0):
    ''' calculates the rotation matrix for rotating samples
            angle_yaw: rotation angle about the +y axis (yaw) [rad]
            angle_pitch: rotation angle about the +x axis (pitch) [rad]
            angle_roll: rotation angle about the +z axis (roll) [rad]
    '''
    R_yaw = np.array([
        [np.cos(angle_yaw), 0, np.sin(angle_yaw)],
        [0, 1, 0],
        [-np.sin(angle_yaw), 0, np.cos(angle_yaw)]])

    R_pitch = np.array([
        [1, 0, 0],
        [0, np.cos(angle_pitch), -np.sin(angle_pitch)],
        [0, np.sin(angle_pitch), np.cos(angle_pitch)]])

    R_roll = np.array([
        [np.cos(angle_roll), -np.sin(angle_roll), 0],
        [np.sin(angle_roll), np.cos(angle_roll), 0],
        [0, 0, 1]])

    R = np.linalg.multi_dot([R_yaw, R_pitch, R_roll])
    return R

def rotate_3D(v, angle_yaw=0, angle_pitch=0, angle_roll=0, R=None):
    ''' calculates the rotation matrix for calculating Bragg plane normal vectors
            v: vector to be rotated
            angle_yaw: rotation angle about the +y axis (yaw) [rad]
            angle_pitch: rotation angle about the +x axis (yaw) [rad]
            angle_roll: rotation angle about the +z axis (yaw) [rad]
            R: rotation matrix, overwrites angle inputs
    '''
    if R is None:
        R = calculate_rotation_matrix(angle_yaw, angle_pitch, angle_roll)
    return np.dot(v, R.T)
```

```python
# x-ray parameter
E = 9.8    # [keV]
lambda0 = 12.39847/E    # [AA]

kin = np.array([ -1398.73758358, -19381.33420562,  45704.44855449]) / 10000
print(kin)
```

```python
def get_pixel_Q(kinVec, x0, y0, z0, shape):
    pixel_pos = np.zeros((shape[0], shape[1], 3))
    pixel_pos[:,:,0] = np.arange(shape[1])[::-1][np.newaxis, :] * 75e-3 + x0
    pixel_pos[:,:,1] = (np.arange(shape[0]))[:, np.newaxis] * 75e-3 + y0
    pixel_pos[:,:,2] = z0

    kout = (pixel_pos / np.linalg.norm(pixel_pos, axis=-1)[:, :, np.newaxis]) * np.linalg.norm(kinVec)
    Q = kout - kinVec[np.newaxis, np.newaxis]
    return np.linalg.norm(Q, axis=-1)
    
```
