```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib as mpl

fig = plt.figure()
ax = fig.add_subplot(111)

tg1 = patches.Rectangle((-5,-50), 10, 150, color="blue", alpha=0.50)
tg2 = patches.Rectangle((-5,-50), 10, 150, color="red",  alpha=0.50)
t1 = mpl.transforms.Affine2D().rotate_deg(0.2) + ax.transData
t2 = mpl.transforms.Affine2D().rotate_deg(-0.2) + ax.transData
tg1.set_transform(t1)
tg2.set_transform(t2)

surface = patches.Rectangle((0,-50), 30, 150, color="green",  alpha=0.50)
surface_t = mpl.transforms.Affine2D().rotate_deg(11.5) + ax.transData
surface.set_transform(surface_t)

ax.add_patch(surface)
ax.add_patch(tg1)
ax.add_patch(tg2)

plt.xlim(-50, 50)
plt.ylim(-75, 75)
ax.set_aspect('equal')
plt.grid(True)

plt.show()
```
