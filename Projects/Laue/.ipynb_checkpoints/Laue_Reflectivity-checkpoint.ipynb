```python
import numpy as np
import crystals
import matplotlib.pyplot as plt
```

```python
# Constant
hbar = 0.0006582119514  # This is the reduced planck constant in keV/fs
c = 299792458. * 1e-9  # The speed of light in um / fs
```

```python
def kev_to_wavevec_length(Ec):
    return Ec/hbar/c
```

```python
def get_reflectivity_curve_Laue(d, Ec, EzCoor, pol, geometry, crystal):

    # Crystal parameters
    k0_crystal = kev_to_wavevec_length(Ec)
    
    wavelength0_crystal = np.pi * 2 / k0_crystal
    dH = crystals.CrystalList[crystal]["spacing"]
    theta = np.arcsin(wavelength0_crystal/2/dH)
    H_len = 2*np.pi/dH     # [1/um]

    # Get coordinates
    Energies = EzCoor + Ec   # keV
    N=EzCoor.shape[0]
    klen_grid = kev_to_wavevec_length(Energies)   # 1/um
    kin = np.concatenate((np.zeros((N,2)), klen_grid[:, np.newaxis]), axis = -1)

    ####### Dynamic Diffraction Equations #######
    chi0 = crystals.CrystalList[crystal]["chi0"]
    if pol == "pi":
        P = np.cos(2*theta)
        chih = crystals.CrystalList[crystal]["chih_pi"]
        chihbar = crystals.CrystalList[crystal]["chihbar_pi"]
    else:
        P = 1
        chih = crystals.CrystalList[crystal]["chih_sigma"]
        chihbar = crystals.CrystalList[crystal]["chih_sigma"]

        
    # get wavevectors
    if geometry == "bragg":
        # bragg
        normal = np.array([0., -np.cos(theta), np.sin(theta)])
    else:
        # laue
        normal = np.array([0, np.sin(theta), np.cos(theta)])  # internal normal to crystal surface

    # Bragg (same as Laue)
    crystal_h = np.array([0,  H_len*np.cos(theta), -H_len*np.sin(theta)])  # internal normal to crystal surface

    dot_kn = np.dot(kin, normal)    # K0.z_hat
    dot_kh = np.dot(kin, crystal_h)
    gamma_0 = np.divide(dot_kn, klen_grid)
    gamma_h = np.divide(dot_kn + np.dot(crystal_h, normal), klen_grid)    
    b = gamma_0/gamma_h
    alpha = np.divide(2*dot_kh+H_len**2, klen_grid**2) 

    LambdaH_s = np.sin(theta)/klen_grid/P/np.sqrt(np.conjugate(chih*chihbar))    # extinction length in symmetric scattering geometry
    LambdaHbar = LambdaH_s*np.sqrt(gamma_0*np.abs(gamma_h))/np.sin(theta)

    ####### R00
    A = d/LambdaHbar    # eq.39 in paper

    # Transmission factor
    C = np.exp(np.divide(1j*chi0*klen_grid*d/2, gamma_0))   # eq.45 in paper

    y_param = klen_grid*LambdaHbar/2/gamma_0*(b*alpha+chi0*(1-b))

    W = np.cos(A/2*np.sqrt(y_param**2+1)) + 1j*y_param*np.sin(A/2*np.sqrt(y_param**2+1))/np.sqrt(y_param**2+1)
    R00 = C*np.exp(-1j*A*y_param/2)*W                   # still flatened array 
    return R00


# --------------------------------------------------------------
#               Recenter R00 position
# --------------------------------------------------------------
def get_laue_transmission(Ec, E0, dEz, R00, eSpectrumInput):
    """
    
    Input:
    R00: reflectivity curve centered at 0 (x-axis: Ec-E0)
    
    """
    Nshift = int((Ec - E0)/dEz)
    R00_shift = np.roll(R00, Nshift)
    #eSpectrumInputIntensity = np.abs(eSpectrumInput)**2

    #eFieldInputIntensity = np.abs(eFieldInput)**2
    eSpectrumTrans = eSpectrumInput*R00_shift
    eFieldTrans = np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(eSpectrumTrans)))    # field of energy shifted Gaussian

    return R00_shift, eFieldTrans, eSpectrumTrans 
```

```python
def get_Ez_coordinate_1D(nEz,dEz, k0=0):
    """Get z-direction coordinate from spectrum"""
    nz = nEz
    #dEz = 2*Ez_range/nEz
    dkz = kev_to_wavevec_length(dEz)

    EzCoor  = np.linspace(start=-nEz * dEz/ 2, stop=nEz * dEz/ 2, num=nEz) # np.arange(nEz) * dEz - nEz * dEz / 2.
    kzCoor = kev_to_wavevec_length(EzCoor) + k0

    zCoor = np.fft.fftshift(np.fft.fftfreq(nz, d=dkz)* 2 * np.pi)
    tCoor = zCoor / c

    return tCoor, zCoor, kzCoor, EzCoor
```

```python
num = 2**10+1

E_center = 9.831             # [keV]
k0 = kev_to_wavevec_length(E_center) # [1/um]

bandwidthZ = 0.2e-3   # [keV] rms
Ez_range = bandwidthZ * 16       # [keV] plotting range from central energy

#nEz = N
dEz = 2 * Ez_range / num

# Get coordinate
tCoor, zCoor, kzCoor, EzCoor = get_Ez_coordinate_1D(num, dEz, k0)
#print(Ez_range*1000, "eV")  
```
